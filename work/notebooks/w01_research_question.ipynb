## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import os, sys, subprocess

IN_COLAB = "google.colab" in sys.modules
REPO_URL = "https://github.com/flyrank-bih/flyrank-ml-internship-starter"
REPO_DIR = "flyrank-ml-internship-starter"

if IN_COLAB:
    if not os.path.isdir(REPO_DIR):
        subprocess.run(["git", "clone", "--depth", "1", REPO_URL, REPO_DIR], check=True)
    os.chdir(REPO_DIR)

import pandas as pd
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

print("Total pages:", df.shape[0])

declining_rate = df["trend_direction"].str.lower().eq("down").mean()
print(f"Declining page rate: {declining_rate:.1%} of all pages are trending down")

visible = df[df["impressions_90d"] >= 100]
ctr_by_type = visible.groupby(["position_tier", "content_type"])["ctr"].mean().sort_values(ascending=True)
print("\nCTR by position tier and content type:")
print(ctr_by_type.round(4).to_string())

stale_visible = df[
    (df["days_since_last_update"] >= 180) &
    (df["impressions_90d"] >= 500)
]
print(f"\nStale visible pages (180+ days, 500+ impressions): {len(stale_visible)} pages")
```
